```python
import pandas as pd

# File paths
file_paths = [ 
    '../dados/python-brasil-2022/inscricoes-palestras.csv', 
    '../dados/python-brasil-2021/inscrições-palestras.csv', 
    '../dados/python-brasil-2020/inscrições-palestras.csv', 
]

dataframes = [pd.read_csv(file_path) for file_path in file_paths]

file_paths = [ 
    '../dados/python-brasil-2023/inscricoes-participantes.csv',
]

dataframes += [pd.read_csv(file_path, delimiter=";") for file_path in file_paths]
df = pd.concat(dataframes, ignore_index=True)
```

```python
df
```

```python

# Function to calculate the percentage of participants for each unique value in a column
def calculate_percentage(column):
    return df[column].value_counts(normalize=True) * 100
```

```python
def normalize_tipo_ingresso(value):
    if type(value) == str:
        value = value.lower()
        if 'inteira' in value:
            return 'presencial'
        elif 'meia' in value:
            return 'presencial'
        elif 'geral' in value:
            return 'presencial'
        elif 'online' in value:
            return 'online'
    return 'outros'


df['tipo_ingresso_normalized'] = df['tipo_ingresso'].apply(normalize_tipo_ingresso)
```

```python
df['tipo_ingresso_normalized']
```

```python
# Calculate the percentage of participants for each column
columns_to_analyze = [
    'tipo_ingresso_normalized',
    'camiseta_tamanho'
]

# Store the results in a dictionary
percentage_results = {}
for column in columns_to_analyze:
    percentage_results[column] = calculate_percentage(column)

# Print the results
for column, percentages in percentage_results.items():
    print("------------------")
    print(f"% participantes para {column}:\n{percentages}\n")
```
